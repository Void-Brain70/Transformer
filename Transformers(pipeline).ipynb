**pipeline() Function**

We will try the pipeline API on is sentiment analysis.It classify texts as positive or negative

**Available pipeline are**

1.feature-extraction (get the vector representation of a text)
2.fill-mask
3.ner (named entity recognition)
4.question-answering
5.sentiment-analysis
6.summarization
7.text-generation
8.translation
9.zero-shot-classification

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I've been waiting for a HuggingFace course my whole life.")
```

```python
# sentiment-analysis
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier(
    ["I've been waiting for a HuggingFace course my whole life.", 
     "I hate this so much!"]
)
```

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier(
    ["I love CSE", 
     "I hate CSE"]
)
```

```python
# the zero-shot-classification pipeline is very powerful
# zero-shot-classification
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```

```python
# zero-shot-classification
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "I am anik chandra,i read cse in university of asia pacific",
    candidate_labels=["anik", "UAP", "CSE"],
)
```

```python
#Text generation
from transformers import pipeline

generator = pipeline("text-generation")
generator("In this course, we will teach you how to")
```

```python
from transformers import pipeline

generator = pipeline("text-generation")
generator("Computer Science and Engineering")
```

```python
# Model-> distilgpt2
# Pipeline-> Text-Generation 
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")
generator(
    "In this course, we will teach you how to",
    max_length=30,
    num_return_sequences=2,
)
```

```python
# Model-> distilgpt2
# Pipeline-> Text-Generation
# "num_return_sequences" and the total length of the output text with the argument max_length 
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")
generator(
    "Bangladesh is a beautiful country",
    max_length=30,
    num_return_sequences=2,
)
```

```python
# Mask Filling
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models.", top_k=2)
```

```python
# Mask Filling
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("Machine learning is a populer now a days <mask> models.", top_k=2)
```

```python
# Mask Filling
# The "top_k" argument controls how many possibilities you want to be displayed.
# The model fills in the special "<mask>" word, which is often referred to as a mask token. 
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("Machine learning is a populer now a days <mask> models.", top_k=3)
```

```python
# Named entity recognition (NER)
# Here the model correctly identified that Sylvain is a person (PER), Hugging Face an organization (ORG), and Brooklyn a location (LOC).
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("My name is Sylvain and I work at Hugging Face in Brooklyn.")
```

```python
 # Named entity recognition (NER)
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("I am a student.My name is anik chandra and I work at Ajaira Company  in Uganda.")
```

```python
# Question answering
# The question-answering pipeline answers questions using information from a given context
from transformers import pipeline

question_answerer = pipeline("question-answering")
question_answerer(
    question="Where do I work?",
    context="My name is Sylvain and I work at Hugging Face in Brooklyn",
)
```

```python
# Question answering
# The question-answering pipeline answers questions using information from a given context
from transformers import pipeline

question_answerer = pipeline("question-answering")
question_answerer(
    question="Where do I work?",
    context="My name is anik chandra and I work at Faltu company in Bangladesh",
)
```

```python
# Translation
from transformers import pipeline

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en")
translator("Ce cours est produit par Hugging Face.")
```

```python
# Translation
from transformers import pipeline

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-bn-en")
translator("কম্পিউটারে বাংলা লেখার অনেক গুলো পদ্ধতি আছে।")
```
